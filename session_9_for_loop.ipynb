**case-3:range(start,stop,step)**

- start value = start

- if step is positive sign

   - it is positive direction 

   - increment by **step vlue**

   - last value = stop-1

- if step is negative sign

    - it is a negative direction (decrement)

    - decrement by **step value**

    - last value = stop+1

```python
for i in range(1,20,2):
    print(i)
```

```python
for i in range(1,20,2):
    print(i,end=' ')
```

```python
for i in range(20,1,-2):
    print(i,end=' ')
```

```python
for i in range(-20,-1,2):
    print(i,end=' ')
```
