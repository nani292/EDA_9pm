**if-elif-else**

- when there is one condition is there 
   - we can expect two answer 
      - in that first answer under if block 
      - second answer under else block
- when there are two answer 
     - we can expect three answer 
        - in that first answer under if block based on condition-1
        - second answer under elif block based on condition-2
        - third answer under else block

```python
if <conditon>:
    <write your code>
elif<condition>:
    <write your code>
else:
    <write your code>
```

```python
# wap ask the user enter a number
# if the number equal to 1 than print one
# otherwise if the number equal to 2 then print two
# otherwise if the number equal to 3 then print three
# otherwise 
n=eval(input('enter a number:'))
if num==1:
    print('one')
elif num==2:
    print('two')
elif==3:
    print('three')
else:
    print('enter a valid number')
```
