**args-kwargs**

- args means we can provide multipule value with out argument name

   - which means while calling function we can provide diractly 'virat',102
 
   - while define the function you no needd to mention argument name also
 
   - the output will comes in **tuple** format : ('viart',102)

- kwargs keyword arguments means we can provide multiple values with argument name

   - while define the function no need to provide the name of the arguments
 
   - while call the functions definitely we need to provide with argument names
 
   - name='virat',runs=102
 
   - the output comes in dict format: {'name':'virat','runs':102} 

```python
def fun1(*args):
    print(args)
fun1(10,20)
```

```python
def fun2(**kwargs):
    print(kwargs)
fun2(a=10,b=20,c=30,d="nani")    
```
