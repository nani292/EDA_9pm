**try-except**

- try except is used to handle the exeption mechanism 

- try except is used to catch and handle the errors 

- when we get the erorr python will stop the execution 

- using try except eventhough we get the error, the execution will not stop

- it will catch all the logical errors automatically 

- the actual code block will write in **try block** 

- is any error occures in try block, will catch the errors in **except** block

- syntax errors will not catch

**try-except-finally**

- original code in **try** block 

- error catch in **except** block 

- **finally** block will activate either or try and except activate 

- assume that try block exceuted then finally block can also exceute

- assume that except block exceuted then aslo finally block exceute

```python
try:
    print('connection astablished')
except Exception as e:
    print(e)
finally:
    print('connection closed')
```
