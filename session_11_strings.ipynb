```python
s1='ola ola ola'
count=0
for i in s1:
    if i=='a':
        count+=1
        print(f'the number of a are {count}')
```

```python
s1='ola ola ola'
count=0
for i in range(s1):
    if i=='a':
        count+=1
        print(f'the number of a are {count}')
```

```python
count=0
for i in range (len(s1)):
    if s1[i]=='a':
        count=count+1
        print(f'the number of a are {count}')
```

```python
sum=0
for i in range(len(s1)):
    if s1[i]=='a':
        sum=sum+i
        print(f'the sum of indexes a are {sum}')
```
