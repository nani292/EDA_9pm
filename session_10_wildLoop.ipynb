- Any loop we required 3 things

   - Intial point (start point)

   - increment or dicrement

   - condition to stop 

- in for loop the above 3 are happens in a single line

- but in wild loop all 3 needs 3 different lines

```python
for i in range(1,10,1):
    print(i)
```

```python
for i in range(10,1,-1):
    print(i)
    i=10
    while i>10:
        print(i)
        i=(i-10,)
```

```python
for i in range(-1,-11,-1):
    print(i,end=' ')
    i=-1
    while i>-1:
        print(i)
        i=i-1
```

```python
for i in range(-10,0,1):
    print(i,end=' ')
    i=-10
    while i>-10:
        print(i)
        i=i-10
```

```python
for i in range(1,11,1):
    print(i,':',end=' ')
    i=1
    while i>1:
        print(i)
        i=i+1
```

```python
for i in range(1,10,1):
    print(i,end=' ')
```

```python
# wap ask the user random number
# take a number from keyboard
# if both are match the won
# otherwise lose 
# give 3 chance
import random
for i in range(3):
    n1=random.randint(1,10)
    n2=int(input('enter the number between 1-10'))
    if n1==n2:
        print('won')
    else:
        print('lost')
```
